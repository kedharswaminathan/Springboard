```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a 
different cost for guests and members!
```

```python
SELECT f.name AS facility_name,
       SUM(CASE 
               WHEN b.memid IS NULL THEN b.slots * f.guestcost
               ELSE b.slots * f.membercost
           END) AS total_revenue
FROM Facilities f
JOIN Bookings b ON f.facid = b.facid
GROUP BY f.name
HAVING SUM(CASE 
               WHEN b.memid IS NULL THEN b.slots * f.guestcost
               ELSE b.slots * f.membercost
           END) < 1000
ORDER BY total_revenue ASC;
```

```python
Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
```

```python
SELECT m.surname, m.firstname,
       r.surname AS recommended_surname,
       r.firstname AS recommended_firstname
FROM Members m
LEFT JOIN Members r ON m.recommendedby = r.memid
ORDER BY m.surname ASC, m.firstname ASC;
```

```python
Q12: Find the facilities with their usage by member, but not guests
```

```python
SELECT f.name AS facility_name,
       COUNT(b.memid) AS member_usage_count
FROM Facilities f
JOIN Bookings b ON f.facid = b.facid
WHERE b.memid IS NOT NULL
GROUP BY f.name
ORDER BY f.name;
```

```python
Q13: Find the facilities usage by month, but not guests
```

```python
SELECT f.name AS facility_name,
       DATE_FORMAT(b.starttime, '%Y-%m') AS booking_month,
       COUNT(b.memid) AS member_usage_count
FROM Facilities f
JOIN Bookings b ON f.facid = b.facid
WHERE b.memid IS NOT NULL
GROUP BY f.name, booking_month
ORDER BY f.name, booking_month;
```
